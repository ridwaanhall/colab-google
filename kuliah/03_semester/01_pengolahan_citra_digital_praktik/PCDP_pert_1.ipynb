```python
import matplotlib.pyplot as plt

#1 pixel itu 100
img = [[200,200,200,25,200],
       [200,200,200,25,200],
       [200,200,200,25,200],
       [200,200,200,25,200],
       [200, 54,200,25,200],
       [200,200,200,25,200],
       [200,200,80, 25,200],
       [200, 80,200,25,200]]

plt.axis('off') #ngehilangin angka
plt.imshow(img,cmap=plt.cm.gray)
#plt.savefig('gambar.png') # karena fig, maka udah gede
plt.imsave('gambar.png', arr=img, cmap='gray') # arr aray matirx, cmap biar warna sesuai
```

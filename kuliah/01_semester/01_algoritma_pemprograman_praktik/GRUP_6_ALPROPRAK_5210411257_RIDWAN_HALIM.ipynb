```python
from google.colab import output

def bank_soal():
  def buat_tabel():
    # library
    import sqlite3
    # membuat koneksi ke dalam database
    conn = sqlite3.connect("DataSoal.db")
    # query sql untuk membuat tabel
    query = '''
    CREATE TABLE banksoal(
      Nomor INT NOT NULL,
      Soal TEXT NOT NULL,
      Jawab1 TEXT NOT NULL,
      Jawab2 TEXT NOT NULL,
      Jawab3 TEXT NOT NULL,
      JawabBenar TEXT NOT NULL
    )
    '''
    # menjalankan query
    conn.execute(query)
    # menutup koneksi
    conn.close()
    print("--BERHASIL MEMBUAT TABEL--")

  def menu():
    print("| PILIH :                           |")
    print("| [0] BUAT TABEL                    |")
    print("| [1] INPUT SOAL DAN JAWABAN        |")
    print("| [2] SOAL                          |")
    print("| [3] LIHAT SEMUA SOAL DAN JAWABAN  |")
    print("| [4] SELESAI                       |")

  def input_soal():
    import sqlite3
    conn = sqlite3.connect("DataSoal.db")
    def masuk():
      print("")
      print('<<<MASUKKAN INPUT SOAL DAN JAWABAN>>>')
      a = int(input("Masukkan nomor soal : "))
      b =     input("Masukkan soal       : ")
      c =     input("Jawaban A           : ")
      d =     input("Jawaban B           : ")
      e =     input("Jawaban C           : ")
      f =     input("Jawaban benar dari A,B,C : ")
      query = f'''
      INSERT INTO banksoal(Nomor, Soal, Jawab1, Jawab2, Jawab3, JawabBenar)
      VALUES (?,?,?,?,?,?)
      '''
      vl = (a,b,c,d,e,f)
      # mengkoneksikan
      conn.execute(query,vl)
      print("--TELAH MENG INPUT--")
    lagi="Y"
    while lagi=="Y" or lagi=="y":
      masuk()
      print()
      lagi=input("Tambah Soal dan jawaban [Y/T] : ")
      print("--SELESAI MENGINPUT--")
    conn.commit()
    conn.close

  def soal_ini():
    # library
    import sqlite3
    import random as ran
    ran.seed()
    # membuat koneksi ke dalam database
    conn = sqlite3.connect("DataSoal.db")
    def soal_soal():
      # query untuk membaca tabel
      query = '''
      SELECT Soal, Jawab1, Jawab2, Jawab3, JawabBenar FROM banksoal
      ORDER BY random()
      '''
      # menjalankan query
      cursor = conn.execute(query)
      # iterasi terhadap cursor
      nilai = 0
      urut = 0
      print("<<<SOAL-SOAL>>>")
      for baris in cursor:
        urut = urut+1
        print(urut,".", baris[0])
        print("   A.",baris[1])
        print("   B.",baris[2])
        print("   C.",baris[3])
        jawab = input("   Jawab [A/B/C] : ")
        jwb = ""
        if jawab == "A" or jawab == "a":
          jwb = baris[1]
        elif jawab == "B" or jawab == "b":
          jwb = baris[2]
        elif jawab == "C" or jawab == "c":
          jwb = baris[3]
        print("")
        print("   yang anda jawab :",jwb)
        print("   yang benar      :",baris[4])
        print("")
        if jwb == baris[4]:
          nilai +=1
      print("total nilai adalah", nilai)
    # tutup koneksi
      conn.close()
    soal_soal()

  def lihat_soal():
    import sqlite3
    # membuat koneksi ke dalam database
    conn = sqlite3.connect("DataSoal.db")
    # query untuk membaca tabel
    query = '''
    SELECT Nomor, Soal, Jawab1, Jawab2, Jawab3, JawabBenar FROM banksoal
    ORDER BY random()
    '''
    # menjalankan query
    cursor = conn.execute(query)
    # iterasi terhadap cursor
    print("<<SEMUA SOAL DAN JAWABAN>>>")
    for baris in cursor:
      print("")
      print("Nomor         =",baris[0])
      print("Soal          =",baris[1])
      print("Jawaban 1     =",baris[2])
      print("Jawaban 2     =",baris[3])
      print("Jawaban 3     =",baris[4])
      print("Jawaban Benar =",baris[5])
    # tutup koneksi
    conn.close()

  def selesai():
    print("<<<TERIMA KASIH>>>")

#########################################
  print ("+===================================+")
  print ("| <<<SELAMAT DATANG DI BANK SOAL>>> |")
  print ("|-----------------------------------|")
  menu()
  print ("|-----------------------------------|")
  pilih = int(input("| Masukkan pilihan : "))
  print ("+===================================+")
  print ("")

  if pilih == 0:
      buat_tabel()
  elif pilih == 1:
      input_soal()
  elif pilih == 2:
      soal_ini()
  elif pilih == 3:
      lihat_soal()
  elif pilih == 4:
    selesai()
  else:
      print ("Ngantuk mas broooo")

lagi="Y"
while lagi=="Y" or lagi=="y":
  output.clear()
  bank_soal()
  print("")
  lagi=input("LAGI [Y/T] : ")
  print("")
```
